

# BAR PLOTTING:

```python



plt.bar(df.Gender.unique(),df.Gender.value_counts())
plt.show()
```

# GIVING NUMBER LOCATION & edgecolor:

```python

plt.bar(df.Gender.unique(),df.Gender.value_counts(),edgecolor='r')
plt.text('male',577,"577")
plt.text('female',314,"314")
plt.show()
```

# DIFFERENT COLORS ON EACH BAR:

```python


plt.bar(df.Gender.unique(),df.Gender.value_counts(),edgecolor='r',color=['k','y'])
plt.text('male',577,"577")
plt.text('female',314,"314")
plt.grid()
```
